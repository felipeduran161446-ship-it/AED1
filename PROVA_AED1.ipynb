```python
while True:
    try:
        idade = int(input("Digite a idade do cliente: "))
        valor_carteira = float(input("Digite o valor total na carteira do cliente (em R$): "))

        if idade < 16:
            print("Entrada permitida apenas com acompanhante.")
        else:
            if valor_carteira >= 30.00:
                print("Ingresso comprado! Bom filme.")
            else:
                print("Saldo insuficiente para o ingresso.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para idade e valor.")

    continuar = input("Deseja verificar outro cliente? (s/n): ").lower()
    if continuar == 'n':
        break

print("Programa encerrado.")
```
