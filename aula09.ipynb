```python
try:
    numero = int(input("Digite um número inteiro:"))
    if numero % 3 == 0 and numero % 5 == 0:
        print(f"O número {numero} é divisível por 3 e por 5 ao mesmo tempo.")
    else:
        print(f"O número {numero} NÃO é divisível por 3 e por 5 ao mesmo tempo,")
except ValueError:
    print("Erro: por favor, digite um número inteiro válido.")
```

```python
def formatar_moeda(valor):
    """Formata o valor no padrão brasileiro"""
    return f"R$ {valor:,.2f}".replace(",", "v").replace(".", ",").replace("v", ".")

def calcular_desconto(valor):
    """Calcula o percentual de desconto baseado no valor"""
    if valor <= 1000:
        return 10
    elif valor <= 5000:
        return 20
    else:
        return 30
        print("="*50)
print("      CALCULADORA DE DESCONTO")
print("="*50)
print("TABELA:")
print("  • Até R$ 1.000,00       → 10% desconto")
print("  • De R$ 1.001,00 a R$ 5.000,00 → 20% desconto")
print("  • Acima de R$ 5.000,00  → 30% desconto")
print("="*50)

try:
    valor = float(input("\nDigite o valor da compra: R$ "))

    if valor < 0:
        print("\n❌ Valor inválido! A compra não pode ter valor negativo.")
    else:
        percentual = calcular_desconto(valor)
        desconto = valor * percentual / 100
        valor_final = valor - desconto

        print("\n" + "─"*50)
        print("           RESULTADO")
        print("─"*50)
        print(f"Valor original:    {formatar_moeda(valor)}")
        print(f"Desconto:          {percentual}%")
        print(f"Valor do desconto: {formatar_moeda(desconto)}")
        print(f"Valor a pagar:     {formatar_moeda(valor_final)}")
        print("─"*50)

except ValueError:
    print("\n❌ Erro! Digite um valor válido (use ponto para separar os centavos).")


```

```python
import math

# Constantes
PE_DIREITO = 2.80
ALTURA_PORTA = 2.10
LARGURA_PORTA = 0.80
RENDIMENTO = 3

print("=== CALCULADORA DE TINTA ===")

try:
    # Entradas
    largura = float(input("Largura do aposento (m): "))
    comprimento = float(input("Comprimento do aposento (m): "))
    capacidade_lata = float(input("Capacidade da lata (litros): "))

    # Cálculos
    area_paredes = 2 * (largura + comprimento) * PE_DIREITO
    area_porta = LARGURA_PORTA * ALTURA_PORTA
    area_pintar = area_paredes - area_porta
    litros_necessarios = area_pintar / RENDIMENTO
    quantidade_latas = math.ceil(litros_necessarios / capacidade_lata)

    # Resultados
    print(f"\nÁrea a pintar: {area_pintar:.2f} m²")
    print(f"Litros necessários: {litros_necessarios:.2f} L")
    print(f"Latas necessárias: {quantidade_latas}")

except:
    print("Erro! Digite valores válidos.")


```
